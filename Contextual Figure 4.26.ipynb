```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file Excel
file_path = r"D:\Semester 4\Metode Peramalan\Harga Beras Validation.xlsx"
df = pd.read_excel(file_path)
```

```python
# Konversi kolom "Date" menjadi tipe data datetime
df['Bulan'] = pd.to_datetime(df['Bulan'])

# Atur kolom "Date" sebagai indeks DataFrame
df.set_index('Bulan', inplace=True)
```

```python
# Plotnya coy
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Harga Beras'], marker='o', linestyle='-')
plt.title('Komposisi Pemetaan Berdasarkan Time Series dari Harga Beras 2021 - 2024')
plt.xlabel('Bulan')
plt.ylabel('Harga Beras')
plt.grid(True)
plt.show()
```
