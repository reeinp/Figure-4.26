```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file Excel
file_path = r"D:\Semester 4\Metode Peramalan\DATASET PENJUALAN BAJU US TABLE 4.10.xlsx"
df = pd.read_excel(file_path)
```

```python
# Konversi kolom "Date" menjadi tipe data datetime
df['Date'] = pd.to_datetime(df['Date'])

# Atur kolom "Date" sebagai indeks DataFrame
df.set_index('Date', inplace=True)
```

```python
# Plotnya coy
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Sales'], marker='o', linestyle='-')
plt.title('U.S. Clothing Sales from January 1992 to December 2003')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
